```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('kc_house_data.csv')
```

```python
raw_csv_data
```

```python
df = raw_csv_data.copy()
```

```python
df.info()
```

```python
df = df.drop(['id'], axis=1)
df.head()
```

```python
df = df.drop(['zipcode'], axis=1)
df.head()
```

```python
df.columns.values
```

```python
column_names_reordered = ['date', 'bedrooms', 'bathrooms', 'sqft_living',
       'sqft_lot', 'floors', 'waterfront', 'view', 'condition', 'grade',
       'sqft_above', 'sqft_basement', 'yr_built', 'yr_renovated', 'lat',
       'long', 'sqft_living15', 'sqft_lot15', 'price']
```

```python
df = df[column_names_reordered]
df.head()
```

```python
df = df.drop(['view'], axis=1)
df.head()
```

```python
# Creating Checkpoint
df_cp = df.copy()
```

```python
type(df_cp['date'][0])
```

```python
df_cp['date'] = pd.to_datetime(df_cp['date'], format = '%Y/%m/%d')
df_cp['date']
```

```python
type(df_cp['date'][0])
```

```python
df_cp.info()
```

```python
df_cp['date'][0]
```

```python
df_cp['date'][0].year
```

```python
list_years = []
```

```python
df_cp.shape
```

```python
for i in range(df_cp.shape[0]):
    list_years.append(df_cp['date'][i].year)
```

```python
list_years
```

```python
len(list_years)
```

```python
df_cp['year'] = list_years
```

```python
df_cp.head()
```

```python
list_months = []
```

```python
for i in range(df_cp.shape[0]):
    list_months.append(df_cp['date'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
df_cp['month'] = list_months
```

```python
df_cp.head()
```

```python
df_cp = df_cp.drop(['date'], axis=1)
df_cp.head()
```

```python
df_cp.columns.values
```

```python
column_names_reordered = ['year', 'month', 'bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors',
       'waterfront', 'condition', 'grade', 'sqft_above', 'sqft_basement',
       'yr_built', 'yr_renovated', 'lat', 'long', 'sqft_living15',
       'sqft_lot15', 'price']
```

```python
df_cp = df_cp[column_names_reordered]
df_cp.head()
```

```python
# Saving data to new csv
df_preprocessed = df_cp.copy()
df_preprocessed.head()
```

```python
df_preprocessed.to_csv('kc_house_data_preprocessed.csv', index=False)
```
