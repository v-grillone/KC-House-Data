```python
import numpy as np
import pandas as pd
```

```python
data_preprocessed = pd.read_csv('kc_house_data_preprocessed.csv')
data_preprocessed.head()
```

```python
data_preprocessed.describe()
```

### Declaring the dependant and independent variables

```python
x = data_preprocessed.iloc[:,:-1]
y = data_preprocessed['price']
```

### Standardizing

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(x)
```

```python
x_scaled = scaler.transform(x)
```

```python
x_scaled
```

### Creating Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(x_scaled, y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

### Creating summary table

```python
x.columns.values
```

```python
reg_summary = pd.DataFrame([['Intercept'], ['year'], ['month'], ['bedrooms'], ['bathrooms'], ['sqft_living'],
       ['sqft_lot'], ['floors'], ['waterfront'], ['condition'], ['grade'],
       ['sqft_above'], ['sqft_basement'], ['yr_built'], ['yr_renovated'], ['lat'],
       ['long'], ['sqft_living15'], ['sqft_lot15']], columns=['Features'])
reg_summary['Coefficients'] = reg.intercept_, reg.coef_[0], reg.coef_[1], reg.coef_[2], reg.coef_[3], reg.coef_[4], reg.coef_[5], reg.coef_[6], reg.coef_[7], reg.coef_[8], reg.coef_[9], reg.coef_[10], reg.coef_[11], reg.coef_[12], reg.coef_[13], reg.coef_[14], reg.coef_[15], reg.coef_[16], reg.coef_[17]
reg_summary
```

### Making predictions

```python
reg.predict(x_scaled)
```

```python
data_predict = data_preprocessed.copy()
```

```python
data_predict['predicted_price'] = reg.predict(x_scaled)
```

```python
data_predict['diff_percent'] = ((data_predict['predicted_price'] - data_predict['price'])/ data_predict['predicted_price']) * 100
```

```python
data_predict = data_predict.iloc[:,-3:]
data_predict.head(25).round(2)
```
